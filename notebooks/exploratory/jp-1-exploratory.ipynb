### columns of note:
 - SalePrice
 - SqFtTotLiving
 - SqFtLot

### ideas
 - proportion of property that house takes up = SqFt1stFloor / SqFtLot

```python
master.SqFtTotLiving = pd.to_numeric(master.SqFtTotLiving)
master.SqFtLot = pd.to_numeric(master.SqFtLot)
master.SqFt1stFloor = pd.to_numeric(master.SqFt1stFloor)
```

```python
master['footprint'] = master.SqFt1stFloor / master.SqFtLot
```

```python
g = sns.regplot(x='SqFtTotLiving', y='SalePrice', data=master)
g.set(ylim=(0, 2000000))
```

```python
f = sns.regplot(x='footprint', y='SalePrice', data=master)
```

```python
df = master_residential[master_residential['SalePrice'] > 0]
```

```python
fig, ax = plt.subplots(figsize=(16,9))
sns.regplot(x='SqFtTotLiving', y='SalePrice', data=df, ax=ax, color='palevioletred')
sns.set_style('darkgrid')
ax.set_xlabel('Square Ft of Living Space')
ax.set_ylabel('Sale Price')
ax.set_title('Relationship Between Sale Price and Square Ft of Living Space')
ax.set_ylim(-1000000,10000000)
fig.savefig('../../src/regression.png')
```

```python
df.corr()
df.shape
```

```python
from statsmodels.formula.api import ols
fsm_df = df[["SalePrice", "SqFtTotLiving"]].copy()
fsm_df.dropna(inplace=True)
formula = 'SalePrice ~ SqFtTotLiving'
fsm = ols(formula=formula, data=fsm_df).fit()
```

```python
fsm.rsquared
```

```python
fsm.params
```

```python
y_hat = fsm.predict()
fsm_resids = fsm.resid
sns.relplot(y_hat, fsm_resids)
```

```python
fsm2_df = df[["SalePrice", "footprint"]].copy()
fsm2_df.dropna(inplace=True)
formula = 'SalePrice ~ footprint'
fsm2 = ols(formula=formula, data=fsm2_df).fit()
```

```python
sns.regplot(x='footprint', y='SalePrice', data=fsm2_df)
```
